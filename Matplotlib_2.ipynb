<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:#00b3e5;"> Matplotlib

```python
import numpy  as np
import pandas as pd
from   matplotlib import pyplot as plt
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Line Plot

```python
x  = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
fig = plt.figure(figsize=(13,5))
```

```python
plt.plot(x,y1,color='g',linestyle=':' ,linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

##### 

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Bar Plot

```python
student = {"Bob":87,"Matt":56,"Sam":27}
```

```python
names  = list(student.keys())
values = list(student.values())
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.5em; color:#00b3e5;"> Vertical

```python
plt.bar(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.5em; color:#00b3e5;"> horizontal

```python
plt.barh(names,values,color='dodgerblue')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
y1=[8,1,7,2,0,3,7,3,2]
y2=[5,8,4,7,1,6,9,1,3]
plt.scatter(x,y1,c="deeppink",marker="o",s=50)
plt.scatter(x,y2,c="dodgerblue",marker="*",s=80)
plt.title("scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Sub Plot

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(2,3,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(2,3,6)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Histogram

```python
data = [1,3,3,3,3,9,9,5,4,4,8,8,8,6,7]
plt.hist(data,color="g",bins=4)
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Box Plot

```python
L1 = [1,5,7,6,9,5,4,5,8,4,5,7,4,5,4,1,2,1,4,1,2,1]
L2 = [9,6,1,4,7,2,5,8,7,8,9,6,5,5,6,8,6,9,5,7,8,5,8,7,7,6,4,1]
L3 = [8,9,1,8,2,7,3,2,2,1,8,2,8,1,2,1,9,8,9,8,2,1,8,3,8,9]

data = [L1,L2,L3]
plt.boxplot(data,showmeans=True,vert=False)
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Violin Plot

```python
L1 = [1,5,7,6,9,5,4,5,8,4,5,7,4,5,4,1,2,1,4,1,2,1]
L2 = [9,6,1,4,7,2,5,8,7,8,9,6,5,5,6,8,6,9,5,7,8,5,8,7,7,6,4,1]
L3 = [8,9,1,8,2,7,3,2,2,1,8,2,8,1,2,1,9,8,9,8,2,1,8,3,8,9]

data = [L1,L2,L3]
plt.violinplot(data,showmedians=True,vert=False) 
plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:1.8em; color:#00b3e5;"> Pie Chart

```python
fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]

plt.pie( quantity,
         labels=fruit,
         autopct='%0.1f%%',
         colors = ['red','orange','yellow','lime'],
         radius=.7
       )
plt.pie([1],colors=['w'],radius=.5)
plt.show()
```
