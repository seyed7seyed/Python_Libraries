#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:#00b3e5;"> plot(x, y)

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x  = np.linspace(0, 10, 50)
y1 = 4 + 2 * np.sin(2 * x)
y2 = 4 - 2 * np.sin(2 * x)
```

```python
# plot
plt.style.use('_mpl-gallery')

fig, ax = plt.subplots( figsize=(7,2) )

ax.plot(x, y1, color='blue', marker='o', markersize=5, linestyle='',  linewidth=2.0)
ax.plot(x, y2, color='red',  marker='',  markersize=5, linestyle='-', linewidth=2.0)

ax.set(title='plot(x,y)', xlabel='X', ylabel='Y',
       xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

ax.grid(False)
plt.show()
```

```python
# plot
plt.style.use('_mpl-gallery')

plt.subplot(2,1,1)
plt.plot(x, y1, color='blue', linestyle='-.', linewidth=2.0) 
plt.title('plot(x,y)')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.xticks(np.arange(1, 8))
plt.yticks(np.arange(1, 8)) 
plt.grid(False)

plt.subplot(2,1,2)
plt.plot(x, y2, color='red',  linestyle='-',  linewidth=2.0)
plt.title('plot(x,y)')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.xticks(np.arange(1, 8))
plt.yticks(np.arange(1, 8)) 
plt.grid(True)

plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:green;"> barbs(X, Y, U, V)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data:
X, Y = np.meshgrid([1, 2, 3, 4], [1, 2, 3, 4])
angle = np.pi / 180 * np.array([[15., 30, 35, 45],
                                [25., 40, 55, 60],
                                [35., 50, 65, 75],
                                [45., 60, 75, 90]])
amplitude = np.array([[5, 10, 25, 50],
                      [10, 15, 30, 60],
                      [15, 26, 50, 70],
                      [20, 45, 80, 100]])
U = amplitude * np.sin(angle)
V = amplitude * np.cos(angle)

# plot:
fig, ax = plt.subplots()

ax.barbs(X, Y, U, V, barbcolor='C0', flagcolor='C0', length=7, linewidth=1.5)

ax.set(xlim=(0, 4.5), ylim=(0, 4.5))

plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:green;"> streamplot(X, Y, U, V)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:#00b3e5;"> Triangular 3D surfaces

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

n_radii = 8
n_angles = 36

# Make radii and angles spaces
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:#00b3e5;"> 3D voxel / volumetric plot

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# Draw cuboids in the top left and bottom right corners
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)

# Combine the objects into a single boolean array
voxelarray = cube1 | cube2

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.voxels(voxelarray, edgecolor='k')

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:#00b3e5;"> 3D wireframe plot

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:violet;"> boxplot( x, y, hue )

```python
import matplotlib.pyplot as plt
#plt.style.use('_mpl-gallery')

import seaborn as sns

fmri = sns.load_dataset('fmri')
fmri.head(3)
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.set(style='whitegrid')

sns_plot = sns.boxplot( x='region', y='signal', data=fmri, hue='event',
                        #order=['frontal', 'parietal'],  # order of x 
                        linewidth=1.0, 
                        palette='Set1' )  # Set1  Set2  Set3  rocket  vlag  blues_d  

sns_plot.set( title="boxplot" )
fig = sns_plot.get_figure() 
#fig.savefig('boxplot.png')
plt.show() 
```

```python
sns.set(rc={'figure.figsize':(13,5)})
sns.set(style='whitegrid')

boxprops    = dict( edgecolor='k', linewidth=1 ) #, facecolor='cyan' 
medianprops = dict( color='k', linewidth=1 )
flierprops  = dict( markersize=7, markerfacecolor='red', markeredgecolor='red', marker='d' ) # 'd', 'x'

sns_plot = sns.boxplot( x='region', y='signal', data=fmri, hue='event',
                        #order=['frontal', 'parietal'],  # order of x 
                        linewidth   = 1.0, 
                        palette     = 'Set1',   # Set1  Set2  Set3  rocket  vlag  blues_d 
                        boxprops    = boxprops,
                        medianprops = medianprops,
                        flierprops  = flierprops  )

sns_plot.set( title="boxplot" )
fig = sns_plot.get_figure() 
#fig.savefig('boxplot.png')
plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:violet;"> countplot( x, hue )

```python
import matplotlib.pyplot as plt
#plt.style.use('_mpl-gallery')

import seaborn as sns

fmri = sns.load_dataset('fmri')
fmri.head(3)
```

```python
sns.set(rc={'figure.figsize':(13,3.5)})
sns.set(style='whitegrid')

#sns_plot = sns.countplot( x=fmri['event'], hue=fmri['region'], palette='Set1' )
sns_plot = sns.countplot( x='event', hue='region', data=fmri, palette='Set1' )
sns_plot.set( title="countplot" )

total = float( fmri.shape[0] )
for p in sns_plot.patches:
    height=p.get_height()
    sns_plot.text(p.get_x()+p.get_width()/2.,height+3,'{0:.0%}'.format(height/total),ha='center')

fig = sns_plot.get_figure() 
#fig.savefig('countplot.png')
plt.show()
```

#  
<span style="font-family:Arial; font-weight:Bold; font-size:2.3em; color:violet;"> heatmap( corr )

```python
import numpy as np

import matplotlib.pyplot as plt
#plt.style.use('_mpl-gallery')

import seaborn as sns

fmri = sns.load_dataset('fmri')
fmri.head(3)
```

```python
sc_cols = ['timepoint','signal']
corr = fmri[sc_cols].corr('pearson')

sns.set(style='whitegrid')
_,ax = plt.subplots(figsize=(3,3))
sns_plot = sns.heatmap( corr, mask=np.zeros_like(corr,dtype=bool),
                        cmap=sns.diverging_palette(220,10,as_cmap=True),
                        square=True, ax=ax, annot=True, center=0.00 )

sns_plot.set( title="heatmap" )
fig = sns_plot.get_figure() 
#fig.savefig('heatmap.png')
plt.show()
```
